```python
import os
import json
import chardet

# Dossier contenant les fichiers JSON
dossier = './json/'  # Change ce chemin selon ton dossier

# Fonction pour détecter l'encodage d'un fichier
def detecter_encodage(fichier):
    with open(fichier, 'rb') as f:
        resultat = chardet.detect(f.read())  # Lire le fichier en mode binaire
    return resultat['encoding']  # Retourne l'encodage détecté

# Fonction pour reformater un fichier JSON
def reformater_fichier_json(fichier_json):
    try:
        # Détecter l'encodage du fichier
        encodage_detecte = detecter_encodage(fichier_json)
        
        if encodage_detecte is None:
            print(f"Impossible de détecter l'encodage du fichier {fichier_json}")
            return
        
        # Ouvrir le fichier avec l'encodage détecté
        with open(fichier_json, 'r', encoding=encodage_detecte, errors='replace') as f:
            donnees = json.load(f)  # Charger le JSON

        # Sauvegarder avec un encodage UTF-8 bien formaté
        with open(fichier_json, 'w', encoding='utf-8') as f:
            json.dump(donnees, f, ensure_ascii=False, indent=4)

        print(f"Fichier bien formaté et converti en UTF-8 : {fichier_json}")

    except json.JSONDecodeE rror:
        print(f"❌ Erreur : Le fichier {fichier_json} n'est pas un JSON valide.")
    except Exception as e:
        print(f"❌ Erreur lors de la modification du fichier {fichier_json}: {e}")

# Parcourir tous les fichiers JSON du dossier
for fichier in os.listdir(dossier):
    if fichier.endswith('.json'):  # Vérifie si c'est bien un fichier JSON
        chemin_fichier = os.path.join(dossier, fichier)
        reformater_fichier_json(chemin_fichier)
```
